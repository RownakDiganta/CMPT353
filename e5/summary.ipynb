```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading data files and creating dataframes from thoses files 
data_1 = pd.read_csv('data-1.csv')
data_2 = pd.read_csv('data-2.csv')
data_3 = pd.read_csv('data-3.csv')
data_4 = pd.read_csv('data-4.csv')
data_5 = pd.read_csv('data-5.csv')
data_6 = pd.read_csv('data-6.csv')
```

```python
# Testing if reading file was successful

print(data_1)
print(data_2)
print(data_3)
print(data_4)
print(data_5)
print(data_6)
```

```python
# In the instruction it is said that we are allowed to use loop. Hence, using it

# Finding
# 1.The mean, standard deviation, and range (min and max) of both the and variables.
# 2.The correlation coefficient (r) between the x and y variables.
# 3.A one sentence description of what's in the data set: 
#   describe the data set to someone and say what you think someone needs to know about it.

data_frames = [data_1,data_2,data_3,data_4,data_5,data_6]

for i in range(len(data_frames)):
    d = data_frames[i]
    print('For data-',(i+1))
    print('Mean of x: ',d['x'].mean())
    print('Mean of y: ',d['y'].mean())
    print('Standard deviation of x: ',d['x'].std())
    print('Standard deviation of y: ',d['y'].std())
    print('Range of x: ',d['x'].min(),'(min) to ',d['x'].max(),'(max)')
    print('Range of y: ',d['y'].min(),'(min) to ',d['y'].max(),'(max)')
    # ref: https://www.geeksforgeeks.org/how-to-calculate-correlation-between-two-columns-in-pandas/ (gained knowledge)
    print('Correlation between x and y:',d['x'].corr(d['y']),'\n')
    #plt.scatter(d['x'],d['y'])
    #plt.show()
    
```
