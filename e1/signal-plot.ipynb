```python
import numpy as np
import matplotlib.pyplot as plt
```

# Creating Fake Sensor Data

```python
N_SAMPLE = 500
```

```python
input_range = np.linspace(0, 2*np.pi, N_SAMPLE, dtype = np.float32) # took 2*np.pi because unit circle ranges from 
                                                                    # 0 to 360 degree
                                                                    # took float32 because the data type might have
                                                                    # decimal points as pi is not a rational number
```

```python
signal = np.sin(input_range)
```

```python
noise = np.random.normal(0,1,N_SAMPLE) # random noise with mu = 0, sigma = 1
                                       # using random.normal for generating noise as mu and sigma is given 
```

```python
assert noise.shape == input_range.shape # Will throw an assertion error if they are not equal 
```

```python
noisy_signal = signal + noise/5 # Shape has to be similar otherwise error.
```

```python
plt.plot(input_range, noisy_signal, "b.", alpha = 0.5)
plt.plot(input_range, signal, "r-",linewidth = 4)      # Trying to plot the signal without any noise
plt.legend(['Sensor readings', 'Truth'])
plt.xlabel('Time')
plt.ylabel('Value of thing we care about')
plt.show()                                             # The plot might vary slightly
```

# Filtering Noise

```python
del signal # with a noisy signal, we don't know the truth: we hope to reconstruct it.
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
filtered = lowess(noisy_signal, input_range, frac = 0.1)
#filtered 2 ta column e print hocche (x,y)
```

```python
plt.plot(input_range, noisy_signal, 'b.', alpha = 0.5)
plt.plot(filtered[:,0], filtered[:,1], "r-", linewidth = 4)
plt.legend(['Sensor readings', 'Reconstructed signal'])
plt.show()
```
